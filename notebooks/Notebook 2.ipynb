```sql
spark.sql("""
SELECT ID, COUNT(*) AS veces
FROM Dim_Clientes_Limpia
GROUP BY ID
HAVING COUNT(*) > 1
""").show()
```

```sql
spark.sql("""
SELECT
  SUM(CASE WHEN ID IS NULL THEN 1 ELSE 0 END) AS id_nulos,
  SUM(CASE WHEN Nombre_completo IS NULL THEN 1 ELSE 0 END) AS nombre_nulos
FROM Dim_Clientes_Limpia
""").show()
```

```sql
spark.sql("""
SELECT Grupo_de_clientes, COUNT(*) AS total
FROM Dim_Clientes_Limpia
GROUP BY Grupo_de_clientes
""").show()
```
